```python
from IPython import get_ipython
from IPython.display import display

data = []

def add_product():
    name = input("Ingresa el nombre del producto: ")
    if any(d == name for d in data):
        print("Este producto ya existe.")
        return
    data.append(name)
    print(f"Producto '{name}' añadido.")

def list_products():
    if not data:
        print("No hay productos en la lista.")
        return
    print("\n--- Lista de Productos ---")
    for index, product in enumerate(data):
        print(f"{index + 1}. Nombre: {product}")

def update_product():
    list_products()
    if not data:
        return
    try:
        index = int(input("Ingresa el número del producto a actualizar: ")) - 1
        if 0 <= index < len(data):
            new_name = input("Ingresa el nuevo nombre del producto: ")
            if any(d == new_name for d in data):
                print("Este producto ya existe.")
                return
            data[index] = new_name
            print("Producto actualizado.")
        else:
            print("Número de producto inválido.")
    except ValueError:
        print("Entrada inválida. Debes ingresar un número.")


def delete_product():
    list_products()
    if not data:
        return
    try:
        index = int(input("Ingresa el número del producto a eliminar: ")) - 1
        if 0 <= index < len(data):
            deleted_product = data.pop(index)
            print(f"Producto '{deleted_product}' eliminado.")
        else:
            print("Número de producto inválido.")
    except ValueError:
        print("Entrada inválida. Debes ingresar un número.")

def main_menu():
    while True:
        print("\n--- Menu ---")
        print("1. Añadir producto")
        print("2. Listar productos")
        print("3. Actualizar producto")
        print("4. Eliminar producto")
        print("5. Salir")

        choice = input("Ingresa tu opción: ")

        if choice == '1':
            add_product()
        elif choice == '2':
            list_products()
        elif choice == '3':
            update_product()
        elif choice == '4':
            delete_product()
        elif choice == '5':
            break
        else:
            print("Opción inválida, intenta de nuevo.")

if __name__ == "__main__":
    main_menu()
```
